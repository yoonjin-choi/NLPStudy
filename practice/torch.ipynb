# PYTORCH

*각 function의 사용법은 documentation [링크 텍스트](https://pytorch.org/docs/stable/generated/torch.nn.MSELoss.html) 참조     
pyTorch에서 제공되는 nn 패키지는 layer들과 거의 동일한 Module의 집합


```
import torch

# N 배치크기, D_in 입력차원
# D_out 출력차원, H는 은닉층의 차원
N,D_in,H,D_out=64,1000,100,10

x=torch.randn(N,D_in) 
y=torch.randn(N,D_out) 

# nn 패키지 활용 - 모델을 순차저 계층 (sequence of layers)
# nn.Sequential : 다른 module을 포함하는 module이며, module을 순차적 적용하여 출력 생성.
model = torch.nn.Sequential(
    torch.nn.Linear(D_in,H),
    torch.nn.ReLU(),
    torch.nn.Linear(H,D_out),
)
```

```
##nn package 에는 손실함수( loss function) 에 대한 정의도 포함.
#여기서는 Mean Squard Error (MSE ) 사용
loss_fn=torch.nn.MSELoss(reduction='sum')

learning_rate=1e-4
```

```
epoch=1000
for t in range(epoch):
  # 순전파 : model에 input x 를 전달하여 output 계산
  y_pred=model(x)

  #손실값을 갖는 Tensor return
  loss = loss_fn(y_pred,y)
  if(t%100 ==99 ):
    print(t,loss.item())
```

```
#Backpropogation 이전 변화도를 0으로
model.zero_grad()

#Model의 학습가능한 모든 매개변수에 대해 손실의 변화도 계산
loss.backward()

#Gradient descent 활용하여 가중치 갱신
with torch.no_grad():
  for params in model.parameters():
    params -= learning_rate*params.grad
```

# Optimizer을 활용하여 가중치 갱신

```
import torch

# N 배치크기, D_in 입력차원
# D_out 출력차원, H는 은닉층의 차원
N,D_in,H,D_out=64,1000,100,10

x=torch.randn(N,D_in) 
y=torch.randn(N,D_out) 

# nn 패키지 활용 - 모델을 순차저 계층 (sequence of layers)
# nn.Sequential : 다른 module을 포함하는 module이며, module을 순차적 적용하여 출력 생성.
model = torch.nn.Sequential(
    torch.nn.Linear(D_in,H),
    torch.nn.ReLU(),
    torch.nn.Linear(H,D_out),
)
```

```

loss_fn=torch.nn.MSELoss(reduction='sum')
```

```
#optim package는 다양한 최적화 알고리즘 포함.
learning_rate = 1e-9
optimizer = torch.optim.Adam(model.parameters(),lr=learning_rate)
```

```
for t in range(500):
  y_pred = model(x)

  loss=loss_fn(y_pred,y)
  if(t%100==99):
    print(t,loss.item())

  # Backpropogation 전 단계에 Optimizer 객체 활용하여 갱신할 parameter에 대한 모든 변화도를 0으로
  # 왜? .backward()호출 시 변화도가 buffer에 누적되기 때문.

  optimizer.zero_grad()

  # backpropogation - loss의 변화도 계산
  loss.backward()
  
  # 매개변수 갱신
  optimizer.step()
```

# pyTorch 사용자 정의 : nn.Module


기존 모듈(sequence)보다 더 복잡한 모델 구성이 필요한 경우   
nn.Module의 서브 클래스로 새 모듈 정의한 후, input을 받아 출력 Tensor을 만든다

```
class TwoLayerNet(torch.nn.Module):
  def __init__(self,D_in,H,D_out):
    #nn.module을 상속받아 생성자에서 linear1, linear2 두 멤버변수 생성
    super(TwoLayerNet,self).__init__()
    self.linear1 = torch.nn.Linear(D_in,H) 
    self.linear2 = torch.nn.Linear(H,D_out)

    def forward(self,x):
      # 입력데이터의 tensor x를 받고 출력데이터의 tensor을 반환
      h_relu=self.linear1(x)
      y_pred=self.linear2(h_relu)

      return y_pred
```

```
N, D_in, H, D_out = 64,1000,100,10

x= torch.randn(N,D_in)
y= torch.randn(N,D_out)
```

```
model = TwoLayerNet(D_in,H,D_out)

criterion = torch.nn.MSELoss(reduction='sum')
optimizer = torch.optim.SGD(model.parameters(),lr=learning_rate)

for t in range(500):
    y_pred = model(x)

    loss=criterion(y_pred,y)
    if(t%100==99):
      print(t,loss.item())

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```
x_train = [[1, 2, 1],
           [1, 3, 2],
           [1, 3, 4],
           [1, 5, 5],
           [1, 7, 5],
           [1, 2, 5],
           [1, 6, 6],
           [1, 7, 7]]
y_train = [2, 2, 2, 1, 1, 1, 0, 0]
```

```
x_train=torch.FloatTensor(x_train)
y_train=torch.LongTensor(y_train)
```

```
class SoftmaxClassifier(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear=nn.Linear(3,3) #4개의 input, 3개에 대한 확률값 - 즉, 3개 중 하나로 분류
  def forward(self,x):
    return self.linear(x)
model = SoftmaxClassifier()
```

```
optimizer=optim.SGD(model.parameters(),lr=0.1)

nb_epochs=1000

for epoch in range(nb_epochs+1):

  prediction = model(x_train)

  cost = F.cross_entropy(prediction,y_train)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epoch % 100 == 0:
        print('Epoch {:4d}/{} Cost: {:.6f}'.format(
            epoch, nb_epochs, cost.item()
        ))
        #print(prediction)
```

```
x_test = torch.FloatTensor([[2,1,1],[3,1,2],[3,3,4]])
y_test = torch.LongTensor([2,2,2])
```

```
###########train function & test function
def train(model,optimizer,x_train,y_train):
  nb_epochs=20
  for epoch in range(nb_epochs) :
      prediction = model(x_train)
      cost = F.cross_entropy(prediction,y_train)

      optimizer.zero_grad()
      cost.backward()
      optimizer.step()

      
      print('Epoch {:4d}/{} Cost: {:.6f}'.format(epoch, nb_epochs, cost.item()))
```

```
optimizer=optim.SGD(model.parameters(),lr=0.1)
model=SoftmaxClassifier()
train(model,optimizer,x_train,y_train)
```

```
def test(model,optimizer,x_test,y_test):
  prediction=model(x_test)
  predicted_class=prediction.max(1)[1] # 확률젤큰값
  correct_count = (predicted_class == y_test).sum().item()
  cost = F.cross_entropy(prediction,y_test)

  print('Acuuracy {} Cost: {:.6f}'.format(correct_count*100/len(y_test), cost.item()))
```

```
test(model,optimizer,x_test,y_test)
```

```
#############training with real data

import numpy as np
train_data=np.loadtxt('/content/drive/MyDrive/test/NLP/diabetes.csv.txt',delimiter=',',dtype=np.float32) #2차원 배열
train_data
x_train=torch.FloatTensor(train_data[:,0:-1])
y_train=torch.FloatTensor(train_data[:,[-1]])
```

```
print(x_train.shape,y_train.shape)
```

```
import torch.nn as nn
class BinaryClassifier(nn.Module): #nn.Module 상속
  def __init__(self):
    super().__init__()
    self.linear=nn.Linear(8,1) #W m * 8(8 dim)
    self.sigmoid=nn.Sigmoid() #b 
  def forward(self,x):
    return self.sigmoid(self.linear(x))
```

```
import torch.optim as optim
import torch.nn.functional as F
import torch
```

```
model = BinaryClassifier()
#optimizer 설정
optimizer=optim.SGD(model.parameters(),lr=1) # model.parameter = W와 B
```

```
nb_epochs =300

for epoch in range(nb_epochs+1):
  #H(x) 계산
  hypothesis=model(x_train)

  #cost 계산 - loss 함수 사용
  cost = F.binary_cross_entropy(hypothesis,y_train)

  #cost통해 H(x) 개선
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  #10번마다 로그 출력
  if(epoch%10==0):
    prediction = hypothesis >= torch.FloatTensor([0.5]) #0.5보다 크면 true 즉 1
    correct_prediction = prediction.float() == y_train #값이 같은지 같으면 true즉 1 -> accuracy 평가
    accuracy = correct_prediction.sum().item() / len(correct_prediction) 
    print('Epoch {:4d}/{} Cost: {:.6f} Accuracy {:2.2f}%'.format(
            epoch, nb_epochs, cost.item(), accuracy * 100,
        ))
    #Cost 감소 Accuracy 증가
```
